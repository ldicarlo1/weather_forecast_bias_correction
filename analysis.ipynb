### 2. Data Preprocessing

```python
# convert observational data to correct units
obs_processed = obs

# convert Kelvin to Celsius | mph to m/s | pascal to mb
obs_processed["temperature_k"] = obs_processed["temperature_k"] - 273.15
obs_processed["dewpoint_k"] = obs_processed["dewpoint_k"] - 273.15
obs_processed["wind_speed_mps"] = obs_processed["wind_speed_mps"] * 0.44704

# rename columns
dict = {"temperature_k":"temperature_c","dewpoint_k":"dewpoint_c","wind_speed_mps":"wind_speed_ms"}
obs_processed = obs_processed.rename(columns=dict)

# remove cloud cover/sfcpress
obs_processed = obs_processed.drop(["sfc_press_pa","cloud_cover"],axis=1)
```

```python
# calculate wind speed for model data
gfs_processed = Preprocessing.calc_windspeed(gfs)
hrrr_processed = Preprocessing.calc_windspeed(hrrr)
ecmwf_processed = Preprocessing.calc_windspeed(ecmwf)

# only select variables of interest
gfs_processed = gfs_processed.drop(["sfc_pressure","sea_level_pressure","cloud_cover"],axis=1)
hrrr_processed = hrrr_processed.drop(["sfc_pressure","sea_level_pressure","cloud_cover"],axis=1)
ecmwf_processed = ecmwf_processed.drop(["sfc_pressure","sea_level_pressure","cloud_cover"],axis=1)
```

```python
# separate into location dataframes in order to merge together with observations
BOS_obs_processed = obs_processed[obs_processed.icao=="KBOS"]
HLC_obs_processed = obs_processed[obs_processed.icao=="KHLC"]

# gfs
BOS_gfs_processed = gfs_processed[gfs_processed.pt=="KBOS"]
HLC_gfs_processed = gfs_processed[gfs_processed.pt=="KHLC"]

#hrrr
BOS_hrrr_processed = hrrr_processed[hrrr_processed.pt=="KBOS"]
HLC_hrrr_processed = hrrr_processed[hrrr_processed.pt=="KHLC"]

# ecmwf
BOS_ecmwf_processed = ecmwf_processed[ecmwf_processed.pt=="KBOS"]
HLC_ecmwf_processed = ecmwf_processed[ecmwf_processed.pt=="KHLC"]
```

### 4. Bias Correction of Forecasts

In order to bias-correct the forecast data, a machine learning algorithm will be implemented. The reason that a machine learning approach will be taken is in my experience of using a simple mean diurnal cycle bias correction (subtracting the mean forecast by the mean observed between hours 0-23), this method is very difficult to scale up. If data is collected on seasonal/yearly basis, then a seasonal bias correction will need to occur, and even perhaps a yearly bias correction. ML algorithms are flexible, easier to scale up, and more often than not highly effective. 

For temperature and wind speed bias correction, I have found that Gradient Boosted Trees or XGBoost perform often better than regular regression techniques. Therefore GBT from Sci-Kit Learn was selected.

###### Experiment

Two variables will be used as inputs, fcst_dt and the forecast variable. The METAR observation variables will be used as the targets. Fcst_dt will represent the temporal variation in the data, particularly how bias increases/decreases as the forecast increases in time. These two variables will be used to train the GBT model. In addition, a bias-correction model will be trained for each variable for each location, and this is to account for the typical differences in bias between two locations (KBOS & KHLC).

My hypothesis is that if GBT is used for bias correction, then temperature/dewpoint/windspeed will all significantly improve in respect to the observational data. However, wind speed will improve the least.

I will first train/test HRRR temperature forecasts for KBOS, then dew point, and windspeed and visualize the results. Then I will perform the training/testing for each model for each location and collect the results.



##### Temperature

```python
# Focus on KBOS first
KBOS_hrrr_analysis_df = hrrr_analysis_df[hrrr_analysis_df.pt=="KBOS"].dropna()
#KBOS_hrrr_analysis_df = KBOS_gfs_analysis_df

# use fcst_dt and model forecasted temperatures as inputs to the model, and the obs data as the target
X_KBOS_hrrr = KBOS_hrrr_analysis_df[["fcst_dt","sfc_air_temperature"]]
Y_KBOS_hrrr = KBOS_hrrr_analysis_df["temperature_c"]

# split into training/testing use 80/20 split
X_train, X_test, y_train, y_test = train_test_split(X_KBOS_hrrr, Y_KBOS_hrrr, test_size=0.2, random_state=42)
```

```python
# train a Gradient Boosted Tree algorithm and print the results
gbr = GradientBoostingRegressor(n_estimators=40)
gbr.fit(X_train,y_train)
pred = gbr.predict(X_test)

# print results
print(f"HRRR Forecast R2 Score: {r2_score(y_test,X_test.sfc_air_temperature)}")
print(f"HRRR Forecast RMSE: {mean_squared_error(y_test,X_test.sfc_air_temperature,squared=False)}")

print(f"Bias Corrected HRRR R2 Score: {r2_score(y_test,pred)}")
print(f"Bias Corrected HRRR RMSE: {mean_squared_error(y_test,pred,squared=False)}")
```

```python
verify_temp = pd.concat([X_test,y_test],axis=1)
verify_temp["pred"] = pred
verify_temp_mean = verify_temp.groupby("fcst_dt").mean()
```

```python
plt.rcParams["figure.figsize"] = (10,5)
plt.plot(verify_temp_mean.index,verify_temp_mean.sfc_air_temperature,label="HRRR Forecast",c="green")
plt.plot(verify_temp_mean.index,verify_temp_mean.pred,label="Bias-Corrected Forecast",c="blue")
plt.plot(verify_temp_mean.index,verify_temp_mean.temperature_c,'--',label="Observed",c="red")
plt.legend()
plt.title("KBOS HRRR Bias-Corrected Temperature Average Performance")
plt.xlabel("Forecast Hour (dt)")
plt.ylabel("Temperature (C)")
plt.xlim([0,48])
plt.savefig("figures/KBOS_HRRR_bias_corrected_temp.png")
```

There is a clear improvement on the forecast as can be seen in the figure above. 

 

Overall excellent results.

###### Overall Improvement

```python
# collect overall improvement by model and method
bias_corrected_results_overall = bias_corrected_results.groupby(["method","model"]).mean()
```

```python
# calculate the % improvement
corrected_rmse = bias_corrected_results_overall.rmse["bias-corrected"]
raw_rmse = bias_corrected_results_overall.rmse["raw forecast"]

perc_improvement =  ((abs(corrected_rmse - raw_rmse)) / raw_rmse) * 100



print(f"Bias Correction of the ECMWF resulted in a {perc_improvement.ECMWF.round(1)}% overall improved forecast")
print(f"Bias Correction of the GFS resulted in a {perc_improvement.GFS.round(1)}% overall improved forecast")
print(f"Bias Correction of the HRRR resulted in a {perc_improvement.HRRR.round(1)}% overall improved forecast")



```

```python
# calculate overall improvement across all models
mean_rmse_bias = bias_corrected_results_overall.rmse["bias-corrected"].mean()
mean_rmse_raw = bias_corrected_results_overall.rmse["raw forecast"].mean()

overall_perc_improvement = (abs(mean_rmse_bias - mean_rmse_raw) / mean_rmse_raw) * 100
print(f"Bias Correction resulted in a {overall_perc_improvement.round(1)}% overall improved forecast")
```
